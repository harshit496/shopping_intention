```python
Revenue_map={False:0,True:1}
dataset['Revenue']=dataset['Revenue'].map(Revenue_map)
Weekend_map={False:0,True:1}
dataset['Weekend']=dataset['Weekend'].map(Weekend_map)
dataset.head()
```

```python
sr=dataset.groupby(['Month'])['Revenue'].sum()
sr=pd.DataFrame(sr)
sr=sr.rename(columns={'Revenue':'true'})
sr['false']=dataset.groupby(['Month'])['Revenue'].count()-(sr.true)
sr['SR']=sr.true/(sr.false+sr.true)
dataset['Month']=dataset['Month'].map(sr['SR'])
```

```python
sr=dataset.groupby(['VisitorType'])['Revenue'].sum()
sr=pd.DataFrame(sr)
sr=sr.rename(columns={'Revenue':'true'})
sr['false']=dataset.groupby(['VisitorType'])['Revenue'].count()-(sr.true)
sr['SR']=sr.true/(sr.false+sr.true)
dataset['VisitorType']=dataset['VisitorType'].map(sr['SR'])
```

```python
#clustering
#administrative dur and bounce rate
x = dataset.iloc[:, [1, 6]].values


x.shape


from sklearn.cluster import KMeans

wcss = []
for i in range(1, 11):
    km = KMeans(n_clusters = i,
              init = 'k-means++',
              max_iter = 300,
              n_init = 10,
              random_state = 0,
              algorithm = 'elkan',
              tol = 0.001)
    km.fit(x)
    labels = km.labels_
    wcss.append(km.inertia_)
    
plt.rcParams['figure.figsize'] = (15, 7)
plt.plot(range(1, 11), wcss)
plt.grid()
plt.tight_layout()
plt.title('The Elbow Method', fontsize = 20)
plt.xlabel('No. of Clusters')
plt.ylabel('wcss')
plt.show()
```

```python
km = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_means = km.fit_predict(x)
plt.scatter(x[y_means == 0, 0], x[y_means == 0, 1], s = 100, c = 'pink', label = 'Cluster1')
plt.scatter(x[y_means == 1, 0], x[y_means == 1, 1], s = 100, c = 'yellow', label = 'Cluster2')
plt.scatter(x[y_means == 2, 0], x[y_means == 2, 1], s = 100, c = 'cyan', label = 'Cluster3')
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:, 1], s = 50, c = 'blue' , label = 'Centroid')

plt.title('Administrative Duration vs Bounce rate', fontsize = 20)
plt.grid()
plt.xlabel('Administrative Duration')
plt.ylabel('Bounce Rates')
plt.legend()
plt.show()
```

```python
kmeans_model = KMeans(n_clusters = 3, random_state = 1).fit(x)

dataset['kmean'] = kmeans_model.labels_
```

```python
x = data1
# removing the target column revenue from x
x = x.drop(['Revenue', 'kmean'], axis = 1)

y = dataset['Revenue']

# checking the shapes
print("Shape of x:", x.shape)
print("Shape of y:", y.shape)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)

# checking the shapes

print("Shape of x_train :", x_train.shape)
print("Shape of y_train :", y_train.shape)
print("Shape of x_test :", x_test.shape)
print("Shape of y_test :", y_test.shape)
```

```python
#comparing models
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scalar = MinMaxScaler()
```

```python
scaled_x_train = pd.DataFrame(scalar.fit_transform(x_train),columns=x_train.columns)
```

```python
scaled_x_test = pd.DataFrame(scalar.transform(x_test),columns=x_test.columns)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report,recall_score, f1_score, precision_score, classification_report
from sklearn.metrics import accuracy_score
knn = KNeighborsClassifier()
knn.fit(scaled_x_train,y_train)
prediction_knn = knn.predict(scaled_x_test)
print("Training Accuracy :", knn.score(scaled_x_train, y_train))
print("Testing Accuracy :", knn.score(scaled_x_test, y_test))
print("The accuracy",accuracy_score(y_test, prediction_knn))

```

```python
print("The accuracy",accuracy_score(y_test, prediction_knn))

print("Recall ",recall_score(y_test, prediction_knn))
print("Precision ",precision_score(y_test, prediction_knn, average="weighted"))
print("f1_score ",f1_score(y_test, prediction_knn, average="weighted"))
print("classification_report",classification_report(y_test,prediction_knn))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score,recall_score, f1_score, precision_score, classification_report
Logistic = LogisticRegression()
Logistic.fit(scaled_x_train,y_train)
prediction_lr = Logistic.predict(scaled_x_test)
```

```python
print("The accuracy",accuracy_score(y_test, prediction_lr))

print("Recall ",recall_score(y_test, prediction_lr))
print("Precision ",precision_score(y_test, prediction_lr, average="weighted"))
print("f1_score ",f1_score(y_test, prediction_lr, average="weighted"))
print("classification_report",classification_report(y_test,prediction_lr))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score, recall_score, f1_score, precision_score, classification_report,confusion_matrix

model = RandomForestClassifier()
model.fit(scaled_x_train, y_train)

y_pred = model.predict(scaled_x_test)


# evaluating the model
print("Training Accuracy :", model.score(scaled_x_train, y_train))
print("Testing Accuracy :", model.score(scaled_x_test, y_test))
```

```python
from sklearn.metrics import confusion_matrix
result = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(result)
```

```python
print("The accuracy",accuracy_score(y_test, y_pred))

print("Recall ",recall_score(y_test, y_pred))
print("Precision ",precision_score(y_test, y_pred, average="weighted"))
print("f1_score ",f1_score(y_test, y_pred, average="weighted"))
print("classification_report",classification_report(y_test,y_pred))
```

```python
print ("the predicted conversion rate is", y_pred.mean())
print ("the actual conversion rate is", y_test.mean())
```
